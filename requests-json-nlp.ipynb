## Question 2

```python
import json
import spacy
from spacytextblob.spacytextblob import SpacyTextBlob

#SpaCy and TextBlob
nlp = spacy.load('en_core_web_sm')
nlp.add_pipe('spacytextblob')

# Read JSON file
with open('lyrics.json', 'r') as json_file:
    lyric_data = json.load(json_file)

# Extract Lyrics
lyrics = lyric_data.get('lyrics', '')

# Sentemint Anaylsis
doc = nlp(lyrics)
polarity = doc._.blob.polarity

# Is Sentiment Postive or Negative?
if polarity > 0:
    sentiment = "Positive"
elif polarity < 0:
    sentiment = "Negative"
else:
    sentiment = "Nuetral"

# Print
print("Lyrics of the song:\n")
print(lyrics)
print("\nPolarity score:")
print(polarity)
print(f"Sentiment Score:  {sentiment}.")
```

3. Write a function that takes an artist, song, and filename, accesses the lyrics.ovh api to get the song lyrics, and writes the results to the specified filename.  Test this function by getting the lyrics to any four songs of your choice and storing them in different files.

## Question 3

```python
import requests
import json

def get_lyrics_and_save(artist, song, filename):
    """
    Fetches lyrics for a given artist and song from the lyrics.ovh API
    and writes the results to a specified file in JSON format.
    
    Parameters:
    - artist (str): The name of the artist.
    - song (str): The title of the song.
    - filename (str): The filename to save the lyrics (JSON format).
    """
    url = f"https://api.lyrics.ovh/v1/{artist}/{song}"
    
    try:
        # Fetch
        response_result = requests.get(url)
        response_result.raise_for_status()  # Raise an error for HTTP issues
        response_result = response_result.json()
        
        # Write Results
        with open(filename, "w") as file:
            json.dump(response_result, file, indent=4)
        
        print(f"Lyrics for '{song}' by {artist} saved to '{filename}'.")
    except requests.exceptions.RequestException as e:
        print(f"Error fetching lyrics for '{song}' by {artist}: {e}")
    except Exception as e:
        print(f"Unexpected error: {e}")

# Song Request
get_lyrics_and_save("Paris Paloma", "Labour", "labour.json")
get_lyrics_and_save("Brandi Carlile", "The Joke", "the_joke.json")
get_lyrics_and_save("The Revivalists", "Soulfight", "soulfight.json")
get_lyrics_and_save("Tracy Chapman", "Fast Car", "fast_car.json")

```

4. Write a function that takes the name of a file that contains song lyrics, loads the file, performs sentiment analysis, and returns the polarity score.  Use this function to print the polarity scores (with the name of the song) of the three files you created in question 3.  Does the reported polarity match your understanding of the song's lyrics?
```python
import json
import spacy
from spacytextblob.spacytextblob import SpacyTextBlob

#spacy and spacytextblob
nlp = spacy.load('en_core_web_sm')
nlp.add_pipe('spacytextblob')

def polarity_lyrics(filename):
    # Load the lyrics data from the file
    with open(filename, "r") as file:
        lyrics_data = json.load(file)
    
    # Extract lyrics
    lyrics = lyrics_data.get("lyrics", "")
    
    # Sentiment Analysis & Polarity
    doc = nlp(lyrics)
    polarity = doc._.blob.polarity
    return polarity

# Test Songs
songs = [
    ("labour.json", "Labour by Paris Paloma"),
    ("the_joke.json", "The Joke by Brandi Carlile"),
    ("soulfight.json", "Soulfight by The Revivalists"),
    ("fast_car.json", "Fast Car by Tracy Chapman")
]

# Print Results
for filename, song_name in songs:
    polarity = polarity_lyrics(filename)
    print(f"Polarity score for '{song_name}': {polarity}")
```
